```python
# PIE CHART - CF_attrition label
plt.figure(figsize=(6,6))
df['CF_attrition label'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Pie Chart - CF_attrition label Distribution')
plt.ylabel('')
plt.show()
```

```python
# BAR CHART - Department
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department', order=df['Department'].value_counts().index)
plt.title('Bar Chart - Count by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# DONUT CHART - Gender
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Donut Chart - Gender Distribution')
plt.show()
```

```python
# Select fewer relevant numerical features
selected_cols = ['Age', 'Monthly Income', 'Total Working Years', 'Years At Company', 'Years In Current Role']
corr_subset = df[selected_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap (Selected Features)')
plt.show()
```

```python
# BOXPLOT - Age by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='Age', data=df)
plt.title('Boxplot - Age by Gender')
plt.show()
```

```python
# HISTOGRAM - Monthly Income
if 'Monthly Income' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['Monthly Income'], kde=True, bins=30)
    plt.title('Histogram - Monthly Income Distribution')
    plt.xlabel('Monthly Income')
    plt.show()
```

```python
# Select top 10 records
top10 = df.head(10)

# Simple line chart of Age vs Monthly Income (top 10 records)
plt.plot(top10['Age'], top10['Monthly Income'], marker='o')
plt.title('Top 10 Records - Age vs Monthly Income')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.grid(True)
plt.show()
```

```python
# Group by 'Years At Company' and calculate mean of selected columns
trend_data = df.groupby('Years At Company')[['Monthly Income', 'Total Working Years', 'Job Satisfaction']].mean().reset_index()

# Line chart for each average trend
plt.figure(figsize=(10, 6))

plt.plot(trend_data['Years At Company'], trend_data['Monthly Income'], marker='o', label='Avg Monthly Income')
plt.plot(trend_data['Years At Company'], trend_data['Total Working Years'], marker='s', label='Avg Total Working Years')
plt.plot(trend_data['Years At Company'], trend_data['Job Satisfaction'], marker='^', label='Avg Job Satisfaction')

# Chart formatting
plt.title('Average Trends by Years At Company')
plt.xlabel('Years At Company')
plt.ylabel('Average Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
